## Creating Tensors

```python
import torch
import matplotlib.pyplot as plt
```

```python
# create a tensor
a = torch.tensor([[1,2,3], [4,5,6]], dtype=torch.int32)
print(a)
```

```python
# ten values evenly distributed between -1 and 1
a = torch.linspace(-1, 1, 10)
print(a)
```

```python
# a tensor of random values 
a = torch.rand(3,4)
print(a)
```

```python
# values normally distributed
a = torch.randn(3,4)
print(a)
```

```python
# random values that are an integer
a = torch.randint(3, 10, size=(3,4))
print(a)
```

```python
# create a zero tensor
a = torch.zeros(2, 3, 4)
print(a)
```

```python
# create a tensor of all 5s
a = torch.full((2,3,4), 5)
print(a)
```

```python
# change the tensor shape
a = torch.randn(3,4,5)
print(a)
```

```python
print(a[1])
```

```python
print(a[1:, 2:4])
```

```python
print(a[:, None, :, None].shape)
```

```python
b = torch.unsqueeze(a, dim=2)
print(a.shape)
print(b.shape)
```

```python
a = torch.randn(3,4)
print(a)
print(a[:, (a > -1).all(axis=0)])
```

```python
# select specific columns
print(a[:, [1,0,0,1]])
```

```python
# convert a 2D tensor into 1D
a = torch.randn(3,4) 
print(a)
print(a.ravel())
```

```python
print(a.reshape(-1))
```

```python
print(a.reshape(3,2,2))
```

```python
print(a.T)
```

```python
# explicitly specify the two axes to swap
print(a.transpose(0, 1))
```

```python
# combine tensors by stacking them (vstack() for vertically and hstack() for horizontally
a = torch.randn(3,3)
b = torch.randn(3,3) 
print(a)
print(b) 
print(torch.vstack([a,b]))
```

```python
c = torch.concatenate([a, b])
print(c)
```

```python
# to split a tensor
print(torch.vsplit(c, 2))
```

```python
# PyTorch tensors can be treated as arrays. So you can often use it in a similar way as NumPy arrays. 
# For example, you have the functions of common mathematical functions:
import torch
a = torch.randn(2,3)
print(a)
print("exp =", torch.exp(a))
print("log =", torch.log(a))
print("sin =", torch.sin(a)) 
print("arctan =", torch.arctan(a)) 
print("abs =", torch.abs(a)) 
print("square =", torch.square(a)) 
print("sqrt =", torch.sqrt(a)) 
print("ceil =", torch.ceil(a)) 
print("round =", torch.round(a)) 
print("clip =", torch.clip(a, 0.1, 0.9))
```

```python
# check if the values of a tensor are nan
b = torch.sqrt(a)
print(b)
print(torch.isnan(b))
```

```python
# Python operators can be applied to the tensors too
a = torch.randn(2, 3) 
b = torch.randn(2, 3) 
print(a)
print(b)
print(a+b)
print(a/b)
print(a ** 2)
```

```python
# matrix multiplications
print(torch.matmul(a, b.T))
print(a @ b.T)
```

```python
# the @ operator from Python can also be used for vector dot-product
a = torch.randn(3)
b = torch.randn(3) 
print(a)
print(b)
print(torch.dot(a, b))
print(a @ b)
```

```python
# find some statistics about the tensor
a = torch.randn(3,4)
print(a)
print(torch.mean(a, dim=0))
print(torch.std(a, dim=0))
print(torch.cumsum(a, dim=0))
print(torch.cumprod(a, dim=0))
```

```python
print(torch.linalg.svd(a))
```

```python
# padding a tensor is done with the following
b = torch.nn.functional.pad(a, (1,1,0,2), value=0)
print(b)
```

```python
## plotting a surface using PyTorch tensors
# create tensors
x = torch.linspace(-1, 1, 100)
y = torch.linspace(-2, 2, 100)
# create the surface
xx, yy = torch.meshgrid(x, y, indexing="xy") # xy-indexing is matching numpy 
z = torch.sqrt(1 - xx**2 - (yy/2)**2)
print(xx)
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([0, 2])
ax.plot_surface(xx, yy, z, cmap="cividis")
ax.view_init(45, 35)
plt.show()
```
